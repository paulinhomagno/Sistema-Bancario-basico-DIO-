```python
from abc import ABC, abstractclassmethod, abstractproperty
from datetime import datetime

class Cliente:

  def __init__(self, endereco):
    self.endereco = endereco
    self.contas = []


  def realizar_transacao(self, conta, transacao):
    transacao.registrar(conta)

  def adicionar_conta(self, conta):
    self.contas.append(conta)

class PessoaFisica(Cliente):
    def __init__(self, nome, dt_nasc, cpf, endereco):
        super().__init__(endereco)
        self.nome = nome
        self.dt_nasc = dt_nasc
        self.cpf = cpf

class PessoaJuridica(Cliente):
    def __init__(self, cnpj, endereco):
        super().__init__(end)
        self.cnpj = cnpj


class Conta:

  def __init__(self, nm_conta, cliente):
    self._saldo = 0
    self._nm_conta = nm_conta
    self._cliente = cliente
    self._historico = Historico()
    self.agencia = "01"

    @classmethod
    def add_conta(cls, cliente, nm_conta ):
        return cls(nm_conta, cliente)

    @property
    def saldo(self):
        return self._saldo

    @property
    def nm_conta(self):
        return self._nm_conta

    @property
    def cliente(self):
        return self._cliente

    @property
    def historico(self):
        return self._historico

    def sacar(self, valor):
        saldo = self._saldo
        excede_saldo = valor > saldo

        if excede_saldo:
          print('Operação negada! Saldo insuficiente!')

        elif valor > 0:
          self._saldo -= valor
          print('Saque realizado com sucesso!')
        else:
          print('Operação inválida!')


    def depositar(self, valor):
        if valor > 0:
            self._saldo += valor
            print('Depósito realizado com sucesso!')
        else:
            print('Operação inválida!')

class ContaCorrente(Conta):
    def __init__(self, numero, cliente, limite_valor = 500, limite_qtde = 3):
        super().__init__(numero, cliente)
        self.limite_valor = limite_valor
        self.limite_qtde = limite_qtde

    def sacar(self, valor):
        qtde_saques = len(
            [transacao for transacao in self.historico.transacoes if transacao["tipo"] == Saque.__name__]
        )

        excedeu_limite = valor > self.limite_valor
        excedeu_saques = qtde_saques >= self.limite

        if excedeu_limite:
            print("Operação negada! Valor excedido.")

        elif excedeu_saques:
            print("Operação negada! Número máximo de saques excedido.")

        else:
            return super().sacar(valor)



class Historico:
    def __init__(self):
        self.transacoes = []

    @property
    def transacoes(self):
        return self._transacoes

    def add_transacao(self, transacao):
        self.transacoes.append(
            {  "tipo": transacao.__class__.__name__,
                "valor": transacao.valor,
                "data": datetime.now().strftime("%d-%m-%Y %H:%M:%S")})



class Transacao(ABC):
  @property
  @abstractproperty
  def valor(self):
    pass


  @abstractclassmethod
  def registrar(self, transacao):
      pass

class Saque(Transacao):
    def __init__(self, valor):
        self._valor = valor

    @property
    def valor(self):
        return self._valor

    def registrar(self, conta):
      transacao = conta.sacar(self.valor)

      if transacao:
        conta.historico.add_transacao(self)


class Deposito(Transacao):

    def __init__(self, valor):
        self._valor = valor

    @property
    def valor(self):
        return self._valor

    def registrar(self, conta):
      transacao = conta.depositar(self.valor)

      if transacao:
        conta.historico.add_transacao(self)
```
